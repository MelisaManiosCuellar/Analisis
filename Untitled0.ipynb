# Correlacion

$$r= \frac{Cov(x,y)}{s_x s_y}$$

donde $s_x$ y $s_y$ son las desviaciones típicas de x y de y. Deben saber que

1. El coeficiente de correlación tiene el mismo signo que la covarianza.
2. El coeficiente de correlación es adimensional: su valor no varía si multiplicamos x por k1 e y por k2, siendo k1 y k2 números no nulos del mismo signo.
3. Si existe una relación lineal exacta entre ambas variables, lo que supone
que todos los puntos deben estar en una línea recta, que podemos escribir como y = a + bx, el coeficiente de correlación es igual a 1 (si b > 0) o –1 (si b < 0).
4. Si no existe una relación lineal exacta (los puntos no están sobre una recta), –1 < r < 1.
